```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def hypothesis(thetas,x):
    return sigmoid(np.dot(thetas.T,x))
```

```python
x = pd.Series(np.arange(-10,10,0.1))
y= x.apply(sigmoid)
plt.plot(x,y)
plt.title("Sigmoid Function")
```

```python
x=np.array(dataset.iloc[:,:-1])
x=np.transpose(x)
y=np.array(dataset.iloc[:,-1])
x = np.insert(x,0,1,axis=0)
```

```python
initial_thetas= np.zeros(x.shape[1])
aplha=0.01
```

```python
def cost_func(thetas,x,y):
    m=x.shape[1]
    summ=-np.dot(y.T,np.log(hypothesis(thetas,x)))-np.dot((1-y).T,np.log(1-hypothesis(thetas,x)))
    return float((1.0/m)*summ)
```

```python
from scipy import optimize
theta=[0,0,0]
def optimizeTheta(mytheta,myX,myy,mylambda=0.):
    result = optimize.fmin(cost_func, x0=mytheta, args=(myX, myy), maxiter=400, full_output=True)
    return result
result=optimizeTheta([0.0,0.0,0.0],x,y)
print result
```
